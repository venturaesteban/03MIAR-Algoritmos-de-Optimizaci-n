## Modelo
- ¿Como represento el espacio de soluciones?
- ¿Cual es la función objetivo?
- ¿Como implemento las restricciones?

Una posible solución podría ser representar el espacio como un vector de 30 posiciones en el que cada posición correspondría a una toma, y el valor que almacenamos en ella nos indicaria que día grabamos esa toma. Por ejemplo, imaginemos que tenemos el siguiente vector $solucion = [2,2,2,1,1,1,3,3,3,...]$, esta representación nos estaría diciendo que las tomas 1, 2 y 3 se grabarían el día 2, que las tomas 4, 5 y 6 se graban el día 1 y las de los días 7, 8 y 9 el día 3, así sucesivamente. Los valores del vector deberían de corresponder con el conjunto de días disponibles, por lo que si tenemos un total de 30 tomas y únicamente podemos grabar 6 por día, tendríamos el cálculo siguiente $\lceil 30/6 \rceil = 5$ que nos indica que necesitamos al menos 5 días. Por tanto, el espacio de soluciones estaría formado por todas las asignaciones posibes de las tomas a los días respetando la restricciónd e máximo 6 por día.
<br>
Teniendo en cuenta que la función objetivo trataría de minimizar el coste total asociado a la contratación de los actores de doblaje, y dado que los actores cobran una cantidad fija por día que acuden al estudio, el coste de un día sería el número de actores distintos que paticiparían en las tomas de ese día. Esto nos lleva a definir la función objetivo como la minimización de la suma de los distintos actores que participan en cada uno de los días de grabación.

$$\min \sum_{d=1}^{D} \text{actores\_distintos}(d)$$

Entendemos $D$ como el número de días y $\text{actores\_distintos}(d)$ cuenta cuántos actores participan en al menos una toma del día $d$.
<br>
Para aplicar las restricciones del problema hay que asegurar dos cosas; la primera es que cada toma se grave exactamente un día y no se pueda ni repetir ni grabar en varios días; la segunda sería que en un mismo día no se puedan grabar más de 6 tomas ya que es la capacidad máxima. Con estas restricciones aseguramos que cualquier solución válida asigne todas las tomas a uno de los días respetando el límite.

## 2. Análisis
- ¿Qué complejidad tiene el problema?
- Espacio de soluciones

Estamos ante un problema de optimización combinatoria ya que básicamente tenemos que repartir 30 tomas en X días de como máximo 6 tomas minimizando lo máximo posible el coste. Esto es equivalente a un problema de particionamiento de conjuntos, es decir, sería un problema de clase NP-Hard en el cual se trata de encontrar una planificación óptima entre un gran número de combinaciones. El orden de complejidad que tendríamos en este problema es:

$$O(5^{30})$$

En este problema, las soluciones crecen de manera exponencial con el número de tomas, lo que hace inviable explorar todas las soluciones mediante fuerza bruta, encontrar una solución óptima usando búsqueda exhaustiva resultaría muy costoso. En este caso, el mejor metodo para resolverlo sería usando algoritmos de búsqueda aproximada o metaheurística como podría ser el recocido simulado o los algoritmos genéticos.
<br>
El espacio de soluciones estaría formado por todas la posibles asignaciones de las 30 tomas a los días de grabación respetando las restricciones comentadas anteriormente. Aunque el límite de 6 tomas por día reduce el número de configuraciones válidas, el número total de posibles planificaciones sigue siendo muy grande, lo que refuerza la necesidad de utilizar métodos de optimización aproximada para abordar el problema.

## 3. Diseño
- ¿Qué técnica utilizo? ¿Por qué?

He decidido usar un recocido simulado para resolver el problema ya que como he comentado anteriormente necesitamos un algoritmo de metaheurística, que es especialmente adecuado para problemas de optimización combinatoria con espacios de búsqueda grandes. El recocido simulado nos permite explorar distintas configuraciones de la planificación y escapar de óptimos locales, ya que como se vio en la pactica AG3 de clase, acepta soluciones peores con una probabilidad controlada por una temperatura que disminuye progresivamente, esto nos permitirá explorar el espacio de soluciones de forma amplia al principio y refinar la solución conforme avanza el proceso.

Se descartan otros algoritmos como el voraz ya que este tipo de método toma decisiones basadas únicamente en la mejor elección local en cada paso. Aunque sea un enfoque rápido y sencillo, no nos garantiza encontrar una buena solución global ya que las decisiones tempranas pueden condicionar negativamente la planificación final.

También he descartado los algoritmos genéticos porque a pesar de que son adecuados para problemas de optimización complejos, implican una mayor complejidad de implementación al requerir definir una población de soluciones, operadores de cruce y mutación, y mecanismos de selección.

### Esquema del algoritmo

- **Generar una solución inicial válida**
- **Calcular el coste de la solución evaluando la función objetivo**
- **Generar una solución vecina**
- **Evaluar la nueva solución**
- **Criterio de aceptación**
    - Si la nueva solución es mejor, se acepta.
    - Si es peor, se acepta con una probabilidad dependiente de la temperatura.
- **Reducir la temperatura**
- **Repetir el proceso**

### 4.1 Carga de datos

### 4.2 Función objetivo

Como hemos visto anteriormente, el coste de la planificación es la suma de actores distintos que tienen que acudir cada día. Para cada día contamos cuántos actores participan en al menos una de las tomas asignadas a ese día.

```python
def actores_del_dia(tomas_del_dia, datos):
    """Devuelve el conjunto de actores que participan en las tomas de un dia"""
    actores = set()
    for toma in tomas_del_dia:
        for actor in range(num_actores):
            if datos[toma][actor] == 1:
                actores.add(actor)
    return actores


def calcular_coste(dias, datos):
    """Calcula el coste total que es la suma de actores distintos por dia"""
    coste = 0
    for tomas_del_dia in dias:
        coste += len(actores_del_dia(tomas_del_dia, datos))
    return coste


def mostrar_planificacion(dias, datos):
    """Muestra la planificacion dia a dia"""
    coste_total = 0
    for i, tomas_del_dia in enumerate(dias):
        actores = actores_del_dia(tomas_del_dia, datos)
        coste_dia = len(actores)
        coste_total += coste_dia
        tomas_mostrar = [t + 1 for t in tomas_del_dia]
        actores_mostrar = [a + 1 for a in sorted(actores)]
        print(f"  Dia {i+1}: Tomas {tomas_mostrar} -> Actores: {actores_mostrar} -> Coste: {coste_dia}")
    print(f"\n  COSTE TOTAL: {coste_total}")
    return coste_total


# Prueba rapida con una asignacion secuencial
dias_prueba = [list(range(i*6, min((i+1)*6, 30))) for i in range(num_dias)]
print(f"Asignacion inicial: {dias_prueba}")
print("Planificacion sin optimizar:")
calcular_coste_prueba = mostrar_planificacion(dias_prueba, datos)
```

### 4.3 Generación de solución inicial

Para poder usar el recocido simulado necesitamos una solución inicial válida, para ello simplemente repartimos las tomas aleatoriamente entre los días disponibles asegurándonos de no superar el límite de 6 tomas por día.

```python
def generar_solucion_inicial(num_tomas, num_dias, max_tomas_dia):
    """Genera una solucion inicial aleatoria pero valida"""
    tomas = list(range(num_tomas))
    random.shuffle(tomas)

    dias = [[] for _ in range(num_dias)]
    for i, toma in enumerate(tomas):
        dias[i % num_dias].append(toma)

    # Verificamos que ningun dia supere el maximo
    for d in dias:
        if len(d) > max_tomas_dia:
            print("ERROR: dia con mas tomas de las permitidas")

    return dias


# Generamos una solucion inicial de ejemplo
solucion_inicial = generar_solucion_inicial(num_tomas, num_dias, max_tomas_dia)

print("Solucion inicial aleatoria:")
coste_inicial = mostrar_planificacion(solucion_inicial, datos)
```

### 4.4 Generación de vecinos

Para generar soluciones vecinas, usamos dos tipos de movimientos:

1. **Movemos una toma:** sacamos una toma de un día aleatorio y la metemos en otro día, siempre que esté libre.
2. **Intercambiar dos tomas:** elegimos dos tomas de días distintos y las intercambiamos entre sí.

Estos movimientos deben mantener las restricciones (máximo 6 tomas por día).

```python
def generar_vecino(dias, max_tomas_dia):
    """Genera una solucion vecina aplicando un movimiento aleatorio"""
    nuevo = copy.deepcopy(dias)
    num_dias = len(nuevo)

    # Elegimos aleatoriamente entre mover o intercambiar
    tipo_movimiento = random.choice(["mover", "intercambiar"])

    if tipo_movimiento == "mover":
        # Elegimos un dia origen que tenga al menos una toma
        dias_con_tomas = [d for d in range(num_dias) if len(nuevo[d]) > 0]
        dia_origen = random.choice(dias_con_tomas)

        # Elegimos un dia destino distinto que tenga hueco
        dias_con_hueco = [d for d in range(num_dias) if d != dia_origen and len(nuevo[d]) < max_tomas_dia]

        if len(dias_con_hueco) == 0:
            # Si no hay hueco en ningun dia, hacemos intercambio en su lugar
            tipo_movimiento = "intercambiar"
        else:
            dia_destino = random.choice(dias_con_hueco)
            # Sacamos una toma aleatoria del dia origen
            i = random.randint(0, len(nuevo[dia_origen]) - 1)
            toma = nuevo[dia_origen].pop(i)
            nuevo[dia_destino].append(toma)

    if tipo_movimiento == "intercambiar":
        dias_con_tomas = [d for d in range(num_dias) if len(nuevo[d]) > 0]
        if len(dias_con_tomas) < 2:
            return nuevo  # no se puede intercambiar

        d1, d2 = random.sample(dias_con_tomas, 2)
        i1 = random.randint(0, len(nuevo[d1]) - 1)
        i2 = random.randint(0, len(nuevo[d2]) - 1)
        # Intercambiamos
        nuevo[d1][i1], nuevo[d2][i2] = nuevo[d2][i2], nuevo[d1][i1]

    return nuevo


# Probamos a generar un vecino
vecino = generar_vecino(solucion_inicial, max_tomas_dia)
print("Vecino generado:")
coste_vecino = mostrar_planificacion(vecino, datos)
```

### 4.5 Recocido simulado

Partimos de una solución inicial y en cada iteración generamos un vecino, si el vecino es mejor, lo aceptamos siempre; si es peor, lo aceptamos con una probabilidad que depende de la diferencia de coste y de la temperatura actual. Bajaremos la temperatura progresivamente por lo que al principio aceptaremos soluciones peores conforme exploremos y al final seremos más selectivos.

```python
#Funcion de probabilidad para aceptar peores soluciones
def probabilidad(T, d):
    if random.random() < math.exp(-1 * d / T):
        return True
    else:
        return False

#Funcion de descenso de temperatura
def bajar_temperatura(T):
    return T * 0.99997


def recocido_simulado(datos, num_dias, max_tomas_dia, TEMPERATURA):
    num_tomas = len(datos)

    solucion_referencia = generar_solucion_inicial(num_tomas, num_dias, max_tomas_dia)
    coste_referencia = calcular_coste(solucion_referencia, datos)

    mejor_solucion = []             #Mejor solucion encontrada
    mejor_coste = 10e100            #Mejor coste encontrado

    N = 0
    historial_costes = []

    while TEMPERATURA > 0.001:
        N += 1

        #Genera una solucion vecina
        vecina = generar_vecino(solucion_referencia, max_tomas_dia)

        #Calcula su coste
        coste_vecina = calcular_coste(vecina, datos)

        #Si es la mejor solucion de todas se guarda (siempre!)
        if coste_vecina < mejor_coste:
            mejor_solucion = copy.deepcopy(vecina)
            mejor_coste = coste_vecina

        #Si la nueva vecina es mejor se cambia
        #Si es peor se cambia segun una probabilidad que depende de T y delta
        if coste_vecina < coste_referencia or probabilidad(TEMPERATURA, abs(coste_referencia - coste_vecina)):
            solucion_referencia = vecina
            coste_referencia = coste_vecina

        #Bajamos la temperatura
        TEMPERATURA = bajar_temperatura(TEMPERATURA)

        if N % 1000 == 0:
            historial_costes.append(coste_referencia)

    print(f"Iteraciones totales: {N}")
    print(f"Mejor coste encontrado: {mejor_coste}")

    return mejor_solucion, mejor_coste, historial_costes


# Ejecutamos el recocido simulado
print("Ejecutando recocido simulado...\n")
mejor_solucion, mejor_coste, historial = recocido_simulado(
    datos, num_dias, max_tomas_dia,
    TEMPERATURA=100
)

print(f"\nPlanificacion obtenida:")
mostrar_planificacion(mejor_solucion, datos)
```

### 4.6 Evolución del coste

Mostramos gráficamente cómo ha ido bajando el coste a lo largo del proceso de enfriamiento para ver la convergencia del algoritmo.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.plot(historial, linewidth=0.8)
plt.xlabel("Iteración (x1000)")
plt.ylabel("Coste")
plt.title("Evolución del coste durante el recocido simulado")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
